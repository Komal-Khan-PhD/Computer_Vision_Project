# Basic functions or manipulation in open cv


```python

import cv2 as cv
from cv2 import resize
import numpy as np
img= cv.imread("resources/image.jpg")
img= cv.resize(img,(500,700))
# resize image
resized_img=cv.resize(img,(450,450))
# convert to gray
gray_img=cv.cvtColor(img, cv.COLOR_BGR2GRAY) #use isColor=false arg in VideoWriter function

# blur the image
blurr_img= cv.GaussianBlur(img,(23,23),0)

#edge detection
edge_img= cv.Canny(img,48,48)

# thickness of lines
mat_kernel= np.ones((7,7),np.uint8)
dilated_img= cv.dilate(edge_img,(7,7),iterations=1) # always keep odd numbers for matrices, iteration means coating

# Make thinner outline
ero_img=cv.erode(dilated_img, mat_kernel,iterations=1)

# Crop image
print("The size of our image is:", img.shape)
crop_img= img[0:200,200:300]

# To show the image in window
cv.imshow("Original image",img)
cv.imshow("Resized image",resized_img)
cv.imshow("Gray image",gray_img)
cv.imshow("blurred image",blurr_img)
cv.imshow("edge detected image",edge_img)
cv.imshow("dilated image",dilated_img)
cv.imshow("Errosion", ero_img)
cv.imshow("Cropped Image", crop_img)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
# Assignment: Define a function to stack multiple images of different sizes and tunes
import cv2 as cv 
import numpy as np

img1=cv.imread("resources\Image_gray.jpg")
img2=cv.imread("resources\image.jpg")

# a function that will take in an image array. If the array is 2D, meaning it's in grayscale, convert it to 3D and return the 3D array. Otherwise, return the original array.
def convert_channels(img):
    if img.ndim == 2:
        return cv.cvtColor(img, cv.COLOR_GRAY2BGR)  
    return img
# a function that will take in a list of image arrays, and return the greatest width and height among the images.
def canvas_size(imgs):
    h_max = 0
    w_max = 0
    for img in imgs:
        h, w, _ = img.shape
        if h > h_max:
            h_max = h
        if w > w_max:
            w_max = w
    return w_max, h_max
#  a function that will take in a list of image arrays, and return a list of images with each image from the original list, only broadcasted to a black canvas. The black canvas should get its dimensions from the canvas_size function
def padded_imgs(imgs):
    max_w, max_h = canvas_size(imgs)
    imgs_new = []
    for img in map(convert_channels, imgs):
        h, w, _ = img.shape
        canvas = np.zeros((max_h, max_w, 3)).astype("uint8")
        canvas[:h, :w] = img
        imgs_new.append(canvas)
    return imgs_new


while True:
    img1, img2=padded_imgs([img1, img2])
    img_stack = np.vstack((img1, img2))
    resized_img=cv.resize(img_stack,(450,450))
    cv.imshow("Stacked Images",img_stack)
    cv.imwrite("resources\img_stacked.jpg",img_stack)
    if cv.waitKey(1) & 0xFF == ord('q'):
            break

cv.destroyAllWindows()
```

# How to change perspective of the image (straighten the document)

```python
import cv2 as cv
import numpy as np

img= cv.imread('resources/warp.jpg')
print(img.shape)
resized_img=cv.resize(img,(450,450))
# Defining points as found in chapter 19
point1= np.float32([[41,31],[14,375],[347,23],[431,355]]) #upper-left, bottom-left, upper-right, bottom-right
width,height= 800,900 
point2= np.float32([[0,0],[0,height],[800,0],[width,height]])
# You can resize the output image 
# out_img= cv.resize
matrix =cv.getPerspectiveTransform(point1,point2)
out_img=cv.warpPerspective(resized_img,matrix,(width,height))

cv.imshow("Original",resized_img)
cv.imshow("Transformed",out_img)
cv.imwrite('resources/warp_perspective.png',out_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# How to detect specific colors inside python

```python
import cv2 as cv
import numpy as np

img=cv.imread('resources/image.jpg')

# Convert into hsv (hue,saturation,value) 
# hsv_img=cv.cvtColor(img,cv.COLOR_BGR2HSV)

# sliders 
def slider():
    pass
path='resources/image.jpg'

cv.namedWindow('Bars')
cv.resizeWindow('Bars',1240,600)

cv.createTrackbar('Hue Min','Bars',0,179,slider)
cv.createTrackbar('Hue Max','Bars',179,179,slider)
cv.createTrackbar('Sat Min','Bars',0,255,slider)
cv.createTrackbar('Sat Max','Bars',255,255,slider)
cv.createTrackbar('Val Min','Bars',0,255,slider)
cv.createTrackbar('Val Max','Bars',255,255,slider)


while True:

    img= cv.imread(path)
    hsv_img=cv.cvtColor(img,cv.COLOR_BGR2HSV)

    hue_min=cv.getTrackbarPos('Hue Min','Bars')
    hue_max=cv.getTrackbarPos('Hue Max','Bars')
    sat_min=cv.getTrackbarPos('Sat Min','Bars')
    sat_max=cv.getTrackbarPos('Sat Max','Bars')
    val_min=cv.getTrackbarPos('Val Min','Bars')
    val_max=cv.getTrackbarPos('Val Max','Bars')

    # to see these changes inside an image
    lower= np.array([hue_min,sat_min,val_min])
    upper= np.array([hue_max,sat_max,val_max])
    mask_img= cv.inRange(hsv_img, lower,upper) 
    out_img= cv.bitwise_and(img,img,mask=mask_img)

    # cv.imshow('Original',img)
    # cv.imshow('HSV',hsv_img)
    cv.imshow('Mask',mask_img)
    cv.imshow('Final Output',out_img)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cv.destroyAllWindows()
```

# Face Detection in images

```python
import cv2 as cv
import numpy as np

face_cascade= cv.CascadeClassifier('resources\haarcascade_frontalface_default.xml')

img=cv.imread('resources\image.jpg')
# print(img.shape) # results in (height, width)
img= cv.resize(img, (912,684)) # divided width and height by 4 to resize 
gray_img= cv.cvtColor(img, cv.COLOR_BGR2GRAY)

faces= face_cascade.detectMultiScale(gray_img,1.1,4)
 
# Draw a rectangle

for (x,y,w,h) in faces:
    cv.rectangle(img, (x,y),(x+w, y+h), (255,0,0), 2)

cv.imshow('image',img)
cv.imwrite('resources/faces.jpg',img)
cv.waitKey(0)
cv.destroyAllWindows()
```
